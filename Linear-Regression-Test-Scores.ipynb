Project Title: Linear Regression for Predicting Test Scores

Introduction: This project aims to build a simple linear regression model to predict students' test scores based on two features: the number of hours studied and the number of hours slept. The goal is to understand the relationship between these variables and how they influence test scores.

Data: The dataset consists of three columns:

Hours Studied: The number of hours a student spent studying.
Hours of Sleep: The number of hours a student slept before the test.
Test Score: The score that the student achieved on the test.
Steps Taken:

Data Understanding:

The data was loaded and examined for any missing values or anomalies. The dataset is small and clean, with no missing values.
Exploratory Data Analysis:

A correlation matrix was generated to explore the relationship between the independent variables (Hours Studied and Hours of Sleep) and the dependent variable (Test Score).
Scatter plots were created to visualize the relationship between the independent variables and the test scores.
Model Building:

The dataset was split into training and testing sets using an 80-20 split.
A linear regression model was trained using the training set.
The model's performance was evaluated using Mean Squared Error (MSE), which was found to be extremely low (2.0194839173657902e-28), indicating high accuracy.
Model Evaluation:

A plot was generated to compare the actual test scores with the predicted test scores. The points were perfectly aligned with the y = x line, indicating that the predictions were nearly identical to the actual values.
Conclusion: The linear regression model demonstrated exceptional accuracy in predicting test scores based on the given features. The low MSE and the perfect alignment of points in the actual vs. predicted plot suggest that the model is well-fitted to the data. However, given the simplicity and small size of the dataset, there is a possibility of overfitting. Further testing on a larger and more complex dataset is recommended to confirm the model's generalizability.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('linear_regression_data.csv')
```

```python
plt.scatter(df['Hours Studied'], df['Test Score'])
plt.title('Relationship between Hours Studied and Test Score')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.show()


```

```python
plt.scatter(df['Hours of Sleep'], df['Test Score'])
plt.title('Relationship between Hours of Sleep and Test Score')
plt.xlabel('Hours of Sleep')
plt.ylabel('Test Score')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df[['Hours Studied', 'Hours of Sleep']]
y = df['Test Score']
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(y_test,y_pred)
plt.title('Actual vs Predicted')
plt.xlabel('Actual Test Score')
plt.ylabel('Predicted Test Score')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')

```
